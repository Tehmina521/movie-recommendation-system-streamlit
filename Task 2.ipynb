# 🎬 Movie Recommendation System using Content-Based Filtering

## 📌 Problem Statement
In modern digital entertainment platforms, users are presented with thousands of movies to choose from. This large volume of content often makes it difficult for users to quickly find movies that match their preferences. Manually browsing through extensive catalogs can be inefficient and time-consuming.  

To address this challenge, recommendation systems are widely used to automatically suggest relevant content to users. A **content-based recommendation system** analyzes the characteristics of items—such as movie genres—and identifies similarities between them. By doing so, it can recommend movies that are closely related to a movie the user already likes.

---

## 🎯 Objective
The objective of this project is to build a **content-based movie recommendation system** that suggests similar movies based on their genre information.

The system performs the following tasks:

- Convert textual movie genre data into numerical feature vectors using **TF-IDF (Term Frequency–Inverse Document Frequency)**.
- Measure similarity between movies using **Cosine Similarity**.
- Recommend the **top 5 most similar movies** based on the selected movie.
- Display recommendations along with their **similarity scores**.
- Develop an **interactive graphical user interface (GUI)** using **Streamlit**, allowing users to enter a movie title and instantly receive personalized recommendations.

This project demonstrates how machine learning techniques can be applied to create an intelligent and interactive recommendation system that improves the user’s content discovery experience.

# Step 1: Import Required Libraries

```python
# Import pandas for handling dataset
import pandas as pd

# Convert text data (genres) into numeric vectors
from sklearn.feature_extraction.text import TfidfVectorizer

# Measure similarity between movies
from sklearn.metrics.pairwise import cosine_similarity
print("✅ Libraries Imported Successfully")
# -------------------------------
# Ignore Warnings
# -------------------------------
import warnings

# Ignore all warnings
warnings.filterwarnings('ignore')

print("✅ All warnings will be suppressed")
```

# Step 2: Load the Movie Dataset

```python
# Load dataset
movies = pd.read_csv("movies.csv")

# Display first 5 rows
print("Dataset shape:", movies.shape)
movies.head()
```

# Step 3: Check Missing Values

```python
movies.isnull().sum()
```

# Step 4: Feature Engineering – Convert Genres to TF-IDF Vectors

```python
# Create TF-IDF object
tfidf = TfidfVectorizer(stop_words='english')

# Transform genres column
tfidf_matrix = tfidf.fit_transform(movies['genres'])

# Check matrix shape
print(tfidf_matrix.shape)
```

# Step 5: Compute Cosine Similarity Matrix

```python
# Calculate cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Print shape
print(cosine_sim.shape)
```

# Step 6: Create Movie Title to Index Mapping

```python
# Create title index mapping
indices = pd.Series(movies.index, index=movies['title']).drop_duplicates()

# Display first few
indices.head()
```

# Step 7: Build the Movie Recommendation Function

```python
def recommend_movies(title, cosine_sim=cosine_sim):

    # Get movie index
    idx = indices[title]

    # Get similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort movies based on similarity
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Select top 5 similar movies (excluding itself)
    sim_scores = sim_scores[1:6]

    # Extract movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Get movie titles
    recommendations = movies.iloc[movie_indices][['title']]

    # Add similarity scores
    recommendations['Similarity Score'] = [i[1] for i in sim_scores]

    return recommendations
```

# Step 8: Test the Recommendation System

```python
recommend_movies("Toy Story (1995)")
```

```python
recommend_movies("Jumanji (1995)")
```

# Step 9: Save the Model for Deployment

```python
import joblib

joblib.dump(cosine_sim, "similarity_model.pkl")
joblib.dump(indices, "movie_indices.pkl")
```

To measure the similarity between movies, **cosine similarity** was applied to the TF-IDF feature matrix. This approach enables the system to identify movies with similar genre characteristics and recommend the **top 5 most relevant movies** along with their similarity scores.
